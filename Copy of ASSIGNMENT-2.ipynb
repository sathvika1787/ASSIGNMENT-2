```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df
```

```python
df.info
```

```python
df.head(10)
```

```python
sns.scatterplot(x="speeding",y="alcohol",data=df)
plt.show
```

```python
#inference: from this plot we can say that there is some positive relationship between alcohol consumption and speeding .Also scatterplot shows some variability in the data , some dots are higher indicating more alcohol consumption while others little less.
```

```python
sns.scatterplot(data=df,x="not_distracted",y="speeding")
```

```python
sns.lineplot(x="speeding",y="alcohol",data=df,ci=None)
```

```python
#inference: As we can see the alcohol consumption for a particular speeding at a pont of time increases and then decreases depending on the speeding.
```

```python
sns.distplot(df['speeding'])
```

```python
#inference:from the graph the most densest range is between 0.15 and 0.20
```

```python
sns.relplot(x="speeding",y="alcohol",data=df,hue="total")
```

```python
#in this relplot we can observe that alcohol increases a speeding increases
```

```python
sns.barplot(data=df,x="total",y="speeding",ci=None)
```

```python
#inference: we see that speeding changes for every value of total, indicating variability.
```

```python
sns.barplot(data=df,x="abbrev",y="speeding",hue="total")
```

```python
#inference: we can observe here in the bar plot that the highest value of speeding is observed for largest value of abbrev
```

```python
sns.countplot(x="abbrev",data=df)
```

```python
sns.jointplot(x="speeding",y="alcohol",data=df)
```

```python
#inference: we can observe a positive relationship in this plot as alcohol increases for increase in speeding .
```

```python
sns.boxplot(x="abbrev",y="speeding",data=df)
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
#inference:effect score closer to 0 says no relationship, closer to 1 or -1 is is positive or negative relationship
```
